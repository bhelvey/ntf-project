```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
import seaborn as sns
```

```python
data = 'data/wageInflation.csv'
df = pd.read_csv(data)
df
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df.dropna(how='any')
```

```python
sns.set(font_scale=1.4)
wage = df.set_index('DATE')['Minimum Wage'].plot(figsize=(12, 10), linewidth= 2.5)
plt.xlabel("Date", labelpad=15)
plt.ylabel("Minimum Wage", labelpad=15)
plt.title("Minimum Wage", y=1.02, fontsize=22)
```

```python
sns.set(font_scale =1.4)
df.set_index('DATE')['Inflation'].plot(figsize=(12, 10), linewidth= 2.5)
plt.xlabel("Date", labelpad=15)
plt.ylabel("Inflation Rate", labelpad=15)
plt.title("Inflation Rate", y=1.02, fontsize=22)
```
