```python
import pandas as pd
import sklearn
from matplotlib import pyplot
from datetime import datetime
from pandas import read_csv
```

```python
def parser(x):
    return datetime.strptime(x,'%Y-%m-%d')

series = read_csv("../data/IndFlaMin.csv", header=0, parse_dates=[1], squeeze=True, date_parser=parser).drop(columns=["Unnamed: 0"])
print(series)
```

```python
dotComCrash = series[(series["Date"] >= "1999-01-01") & (series["Date"] < "2003-03-01")] 
```

```python
dotComCrash = dotComCrash.set_index(["Date"])
```

```python
dotComCrash.plot()
```

```python
fc2008 = series[(series["Date"] >= "2007-01-01") & (series["Date"] < "2009-06-01")] 
```

```python
fc2008 = fc2008.set_index(["Date"])
```

```python
fc2008.plot()
```
